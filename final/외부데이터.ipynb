```python
import pandas as pd

df.rename(columns={
    '기준일로부터 향후 1년내 부도, 기업회생, 90일이상 금융연체, 대지급 등 발생여부(신용정보원 기준). 1: 향후1년내 부도발생, 0: 향후1년내 부도미발생': 'target'
}, inplace=True)
print(df.shape)              # (124989, 53)
print(df['target'].value_counts())  # 0과 1 값 개수
print("Missing values:", df.isnull().sum().sum())  # 0이면 결측치 없음
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier

# 드롭할 컬럼(ID, 날짜, 타깃 제외)
drop_cols = ['target', '기준년월', '가명 식별자']
df_model = df.drop(drop_cols, axis=1)

# 문자열 타입 컬럼 자동 탐색
cat_cols = df_model.select_dtypes(include=['object']).columns.tolist()

# One-hot 인코딩
df_model = pd.get_dummies(df_model, columns=cat_cols, drop_first=True)

# 5) 피처·레이블 분리
X = df_model
y = df['target']

# 6) 학습/테스트 분할 (stratify)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, stratify=y, test_size=0.2, random_state=42
)

# 7) 모델 선언 (클래스 불균형 고려)
log_clf = LogisticRegression(max_iter=1000, class_weight='balanced')
rf_clf  = RandomForestClassifier(n_estimators=100, class_weight='balanced', random_state=42)
scale = (y_train==0).sum()/(y_train==1).sum()
xgb_clf = XGBClassifier(
    n_estimators=100,
    scale_pos_weight=scale,
    use_label_encoder=False,
    eval_metric='logloss',
    random_state=42
)

# 8) 모델 학습
log_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
xgb_clf.fit(X_train, y_train)
```

```python
import numpy as np

# 학습/테스트 세트에서 inf, -inf 개수 확인
print("Train inf count:\n", np.isinf(X_train).sum())
print("Test inf count:\n",  np.isinf(X_test).sum())

# 무한대 → NaN 으로 변환
X_train.replace([np.inf, -np.inf], np.nan, inplace=True)
X_test.replace([np.inf, -np.inf], np.nan, inplace=True)
```

```python
# NaN이 있는 행 제거
mask = ~X_train.isnull().any(axis=1)
X_train, y_train = X_train[mask], y_train[mask]

# 테스트도 동일하게
mask_test = ~X_test.isnull().any(axis=1)
X_test, y_test  = X_test[mask_test], y_test[mask_test]

# 8) 모델 학습
log_clf.fit(X_train, y_train)
rf_clf.fit(X_train, y_train)
xgb_clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, recall_score

for model, name in [(log_clf, "Logistic"), (rf_clf, "RandomForest"), (xgb_clf, "XGBoost")]:
    y_pred = model.predict(X_test)
    cm = confusion_matrix(y_test, y_pred)
    print(f"\n{name} Confusion Matrix:\n", cm)
    print(f"{name} Recall: {recall_score(y_test, y_pred):.3f}")
```
