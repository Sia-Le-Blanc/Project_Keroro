```python
import requests
import pandas as pd
import xml.etree.ElementTree as ET

# 1) API 설정
API_KEY = "hloQJu4UZs7hPIPUkLWyZI4vv20axEsPwEUSobtMB1xiBinkDR1m86Cf7DzMvlcGvXNXgN45kUuHK0Att3Ms2g%3D%3D"  # 발급받은 서비스 키로 교체하세요
API_URL = "https://apis.data.go.kr/1613000/pubStorgeOpenApiService/getMolitStatList"

# 2) 전체 데이터 요청 (numOfRows를 충분히 크게 설정)
params = {
    "serviceKey": API_KEY,
    "pageNo": 1,
    "numOfRows": 1000,      # 총 레코드 수 이상으로 설정
    "returnType": "XML"
}

resp = requests.get(API_URL, params=params)
resp.raise_for_status()

# 3) XML 파싱
root = ET.fromstring(resp.content)
items = root.find(".//items")

# 4) 레코드 추출 및 분양률 계산
records = []
for item in items.findall("item"):
    complex_name = item.findtext("pjtname", default="Unknown")       # 단지명 태그
    total_units = int(item.findtext("suplyHshldCo", default="0"))     # 총분양가구수 태그
    unsold_units = int(item.findtext("unsgndHshldCo", default="0"))   # 당월미분양가구수 태그
    sale_rate = (total_units - unsold_units) / total_units * 100 if total_units > 0 else None
    records.append({
        "단지명": complex_name,
        "총분양가구수": total_units,
        "당월미분양가구수": unsold_units,
        "분양률(%)": round(sale_rate, 2) if sale_rate is not None else None
    })

# 5) DataFrame 생성 및 표시
df = pd.DataFrame(records)
```
