```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose

# 2) 기본 정보 출력
print("=== DataFrame Info ===")
print(df.info(), "\n")
print("=== Summary Statistics ===")
print(df.describe(include='all').transpose(), "\n")
print("=== Missing Values ===")
print(df.isnull().sum(), "\n")

# 3) 범주형 변수 분포 시각화
categorical_cols = ['업종_대분류', '업종_중분류', '법인_고객등급']
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    df[col].value_counts().plot(kind='bar')
    plt.title(f'{col} 분포')
    plt.xlabel(col)
    plt.ylabel('건수')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()

# 4) 수치형 변수 상관관계 히트맵
num_cols = df.select_dtypes(include=['int64', 'float64']).columns
corr = df[num_cols].corr()

plt.figure(figsize=(12, 10))
plt.imshow(corr, cmap='coolwarm', interpolation='nearest', aspect='auto')
plt.colorbar(label='상관계수')
plt.xticks(range(len(num_cols)), num_cols, rotation=90)
plt.yticks(range(len(num_cols)), num_cols)
plt.title('수치형 변수 상관관계 히트맵')
plt.tight_layout()

# 5) 시계열 분해 예시: 기준년월별 요구불예금잔액 합계
ts = df.groupby('기준년월')['요구불예금잔액'].sum()
ts.index = pd.to_datetime(ts.index.astype(str), format='%Y%m')

decomp = seasonal_decompose(ts, model='additive', period=12)
fig = decomp.plot()
fig.set_size_inches(10, 8)
plt.tight_layout()

plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier, export_text
import pandas as pd


# 2) 학습용 데이터 준비
X = df.drop(columns=['기준년월', '업종_대분류','업종_중분류','사업장_시도','사업장_시군구','법인_고객등급'])
X = X.select_dtypes(include=['float64','int64']).fillna(0)
y = df['전담고객여부'].map({'Y':1,'N':0})

# 3) 결정트리 모델 학습
tree = DecisionTreeClassifier(max_depth=3)
tree.fit(X, y)

# 4) 트리 구조 출력
print(export_text(tree, feature_names=list(X.columns)))
```

```python
# 전담고객여부 Y인 사람의 여신_운전자금대출잔액 최댓값
max_Y = df.loc[df['전담고객여부'] == 'Y', '여신_운전자금대출잔액'].max()

# 전담고객여부 N인 사람의 여신_운전자금대출잔액 최솟값
min_N = df.loc[df['전담고객여부'] == 'N', '여신_운전자금대출잔액'].min()

print(f"전담고객여부 = 'Y'인 경우 여신_운전자금대출잔액 최댓값: {max_Y} 백만원")
print(f"전담고객여부 = 'N'인 경우 여신_운전자금대출잔액 최솟값: {min_N} 백만원")
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.tree import DecisionTreeClassifier, export_text, plot_tree
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt

# 1) 데이터 로드 및 전처리
X = df.drop(columns=['기준년월', '업종_대분류', '업종_중분류', '사업장_시도', '사업장_시군구', '법인_고객등급', '전담고객여부'])
X = X.select_dtypes(include=['float64', 'int64']).fillna(0)
y = df['전담고객여부'].map({'Y': 1, 'N': 0})

# 2) 학습·검증 데이터 분할
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, stratify=y, random_state=42
)

# 3) 하이퍼파라미터 그리드 서치 (max_depth, min_samples_leaf)
param_grid = {
    'max_depth': [3, 4, 5, 6],
    'min_samples_leaf': [50, 100, 200]
}
grid = GridSearchCV(
    DecisionTreeClassifier(random_state=42),
    param_grid,
    cv=5,
    scoring='accuracy'
)
grid.fit(X_train, y_train)
best_tree = grid.best_estimator_

# 4) 최적 파라미터 및 트리 규칙 출력
print("Best Parameters:", grid.best_params_)
print("\n=== Decision Tree Rules ===")
print(export_text(best_tree, feature_names=list(X.columns)))

# 5) 테스트 세트 성능 평가
y_pred = best_tree.predict(X_test)
print("\n=== Classification Report ===")
print(classification_report(y_test, y_pred, target_names=['N','Y']))

# 6) 혼동 행렬 시각화
cm = confusion_matrix(y_test, y_pred)
fig, ax = plt.subplots(figsize=(4, 4))
ax.matshow(cm)
for i in range(cm.shape[0]):
    for j in range(cm.shape[1]):
        ax.text(j, i, cm[i, j], ha='center', va='center')
ax.set_xticks([0, 1]); ax.set_xticklabels(['N', 'Y'])
ax.set_yticks([0, 1]); ax.set_yticklabels(['N', 'Y'])
ax.set_xlabel('Predicted'); ax.set_ylabel('Actual')
ax.set_title('Confusion Matrix')
plt.tight_layout()
plt.show()
```
