```python
# 업종별 평균 계산
grouped = df.groupby('업종(대분류)')[['부채비율', '기준일로부터 향후 1년내 부도, 기업회생, 90일이상 금융연체, 대지급 등 발생여부(신용정보원 기준). 1: 향후1년내 부도발생, 0: 향후1년내 부도미발생']].mean().reset_index()

# 컬럼명 정리
grouped.columns = ['업종', '부채비율', '부도위험률']

# 시각화
plt.figure(figsize=(10, 6))
sns.scatterplot(data=grouped, x='부채비율', y='부도위험률', hue='업종', s=100)
plt.title('업종별 부채비율 vs 부도위험률')
plt.xlabel('부채비율 (%)')
plt.ylabel('부도위험률')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 부도 여부 컬럼명 정리
target_col = '기준일로부터 향후 1년내 부도, 기업회생, 90일이상 금융연체, 대지급 등 발생여부(신용정보원 기준). 1: 향후1년내 부도발생, 0: 향후1년내 부도미발생'

# ✅ 1. 업종별 영업이익율 vs 부도위험률
industry_profit_risk = df.groupby('업종(대분류)')[['영업이익율', target_col]].mean().reset_index()
industry_profit_risk.columns = ['업종', '영업이익율', '부도위험률']

plt.figure(figsize=(12, 6))
sns.scatterplot(data=industry_profit_risk, x='영업이익율', y='부도위험률', hue='업종', s=120)
plt.title('업종별 영업이익율 vs 부도위험률')
plt.xlabel('영업이익율 (%)')
plt.ylabel('부도위험률')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# ✅ 2. 시도명별 부채비율 vs 부도위험률
region_debt_risk = df.groupby('시도명')[['부채비율', target_col]].mean().reset_index()
region_debt_risk.columns = ['시도명', '부채비율', '부도위험률']

plt.figure(figsize=(12, 6))
sns.scatterplot(data=region_debt_risk, x='부채비율', y='부도위험률', hue='시도명', s=120)
plt.title('시도명별 부채비율 vs 부도위험률')
plt.xlabel('부채비율 (%)')
plt.ylabel('부도위험률')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()

# ✅ 3. 부도위험률 높은 업종 Top 10
top_risk_industries = industry_profit_risk.sort_values(by='부도위험률', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(data=top_risk_industries, x='부도위험률', y='업종', palette='Reds_r')
plt.title('부도위험률 상위 10개 업종')
plt.xlabel('부도위험률')
plt.ylabel('업종')
plt.tight_layout()
plt.show()
```

```python
# 부채비율 이상치 제거 (예: 음수 또는 너무 큰 값 제거)
clean_region_df = region_debt_risk[(region_debt_risk['부채비율'] > 0) & (region_debt_risk['부채비율'] < 1000)]

plt.figure(figsize=(12, 6))
sns.scatterplot(data=clean_region_df, x='부채비율', y='부도위험률', s=100, color='skyblue')

# 시도명 텍스트 추가
for i in range(len(clean_region_df)):
    plt.text(
        x=clean_region_df['부채비율'].iloc[i],
        y=clean_region_df['부도위험률'].iloc[i],
        s=clean_region_df['시도명'].iloc[i],
        fontsize=9,
        ha='center'
    )

plt.title('시도명별 부채비율 vs 부도위험률', fontsize=14)
plt.xlabel('부채비율 (%)')
plt.ylabel('부도위험률')
plt.grid(True)
plt.tight_layout()
plt.show()
```
