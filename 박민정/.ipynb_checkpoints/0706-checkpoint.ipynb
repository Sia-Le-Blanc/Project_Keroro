```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
# 모든 열 보기 설정
pd.set_option('display.max_columns', None)
# ✅ 한글 폰트 설정 (예: 나눔바른고딕 사용)
plt.rcParams['font.family'] = 'NanumBarunGothic'
plt.rcParams['axes.unicode_minus'] = False  # 마이너스 깨짐 방지
df = pd.read_csv('./(아이엠뱅크) 2025 교육용 데이터.csv',encoding='cp949')
df
```

```python
df.info()
```

```python
# 고객등급별 전담고객 여부 분포표
cross_tab = pd.crosstab(df['법인_고객등급'], df['전담고객여부'], normalize='index') * 100
cross_tab = cross_tab.round(1)
cross_tab
```

```python
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import platform

# ✅ 한글 폰트 설정
if platform.system() == 'Windows':
    plt.rc('font', family='Malgun Gothic')
elif platform.system() == 'Darwin':
    plt.rc('font', family='AppleGothic')
else:
    plt.rc('font', family='NanumGothic')
plt.rcParams['axes.unicode_minus'] = False

# ✅ 기존 코드
cross_tab.plot(kind='bar', stacked=True, color=['lightgray', 'steelblue'], figsize=(7,4))
plt.title('고객등급별 전담고객 여부 비율')
plt.xlabel('법인 고객등급')
plt.ylabel('비율 (%)')
plt.legend(title='전담고객여부', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
# 1️⃣ 지역별 전담고객 비율 분석
지역별_비율 = pd.crosstab(df['사업장_시도'], df['전담고객여부'], normalize='index') * 100
지역별_비율 = 지역별_비율.round(1).sort_values(by='Y', ascending=False)
지역별_비율
```

```python
# 시각화
plt.figure(figsize=(10, 6))
지역별_비율.plot(kind='bar', stacked=True, color=['lightgray', 'steelblue'])
plt.title('지역별 전담고객 여부 비율')
plt.xlabel('사업장 시도')
plt.ylabel('비율 (%)')
plt.legend(title='전담고객여부')
plt.xticks(rotation=45)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

```python
# 2️⃣ 업종별 전담고객 비율 분석
업종별_비율 = pd.crosstab(df['업종_대분류'], df['전담고객여부'], normalize='index') * 100
업종별_비율 = 업종별_비율.round(1).sort_values(by='Y', ascending=False)
업종별_비율
```

```python
plt.figure(figsize=(14, 6))  # 가로 폭 증가
업종별_비율.plot(kind='bar', stacked=True, color=['lightgray', 'steelblue'])
plt.title('업종별 전담고객 여부 비율')
plt.xlabel('업종 대분류')
plt.ylabel('비율 (%)')
plt.legend(title='전담고객여부')
plt.xticks(rotation=45, ha='right', fontsize=9)  # ← 핵심 수정
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.show()
```

- 전담 vs 비전담 고객군에서 예대비율의 평균 차이
- 예대비율 높은 비전담 고객 → 전담 전환 필요 대상
- 예금 많고 여신 적은 고객 → 여신 확대 대상

- 예금/여신 많지만 거래점수가 낮은 고객 필터링
- 특히 전담이 아닌 고객 중 우량 고객 → RM 마케팅 타깃

- 업종_대분류 기준으로 예대비율 평균 계산
- 특정 업종에서 예대비율이 100% 이상 → 리스크 집중
- 여신에만 치우친 업종 → 리스크 관리 전략 필요

1. 업종별 기업여신 집중도 분석
목적: 부동산 PF 리스크처럼, 특정 업종에 여신이 과도하게 집중된 기업고객군이 있는지 파악

주요 변수:

업종_대분류, 업종_중분류

여신한도금액, 여신 운전자금 대출잔액, 여신 시설자금 대출잔액

분석:

업종별 총 여신 금액 집계

업종별 평균 여신 / 예대비율

전담고객 여부에 따른 차이 분석

2. 우량 기업 고객의 전담 여부 & 누락 고객 탐색
목적: 여신 잔액은 큰데 전담이 아닌 고객 → RM이 놓치고 있는 핵심 기업

주요 변수:

법인_고객등급, 전담고객여부, 총여신, 총예금

분석:

전담/비전담 고객 비교

비전담인데 여신 규모 큰 기업 리스트업

비전담 우수등급 고객 비율 파악

3. 지역별 기업 고객 기반 분석
목적: PRM/RM 배치 전략 수립: 여신 실적 많은 지역 + 비전담 많음 → 인력 확충 필요

주요 변수:

사업장_시도, 사업장_시군구, 총여신, 전담고객여부

분석:

지역별 총 여신 금액

전담고객 비율

지역별 예대비율



```python
# 1. 전담 vs 비전담 고객군 예대비율 평균 비교
df['총여신'] = df[['여신한도금액', '여신_운전자금대출잔액', '여신_시설자금대출잔액']].sum(axis=1)
df['총예금'] = df[['요구불예금잔액', '거치식예금잔액', '적립식예금잔액']].sum(axis=1)
df['예대비율'] = df['총여신'] / df['총예금'].replace(0, pd.NA)

전담별_예대비율 = df.groupby('전담고객여부')['예대비율'].mean().reset_index()
전담별_예대비율
```

```python
# 2. 예대비율 높은 비전담 고객 리스트업 → 전담 전환 필요 대상
비전담_고객 = df[df['전담고객여부'] == 'N'].copy()
비전담_고객 = 비전담_고객.sort_values('예대비율', ascending=False)
비전담_고객_상위 = 비전담_고객[['법인_고객등급', '업종_대분류', '사업장_시도','총여신', '총예금', '예대비율']].head(10)
비전담_고객_상위
```

```python
# 3. 예금 많고 여신 적은 고객 → 여신 확대 대상
df['여신/예금 비율'] = df['총여신'] / df['총예금'].replace(0, pd.NA)
예금_많고_여신_적은 = df[df['여신/예금 비율'] < 1].copy()
예금_많고_여신_적은 = 예금_많고_여신_적은.sort_values('총예금', ascending=False)
예금_많고_여신_적은_상위 = 예금_많고_여신_적은[['법인_고객등급', '업종_대분류', '사업장_시도','총여신', '총예금', '여신/예금 비율']].head(10)
예금_많고_여신_적은_상위
```

```python
import numpy as np
# 총여신, 총수신 계산
df['여신합계'] = df[['여신한도금액', '여신_운전자금대출잔액', '여신_시설자금대출잔액']].sum(axis=1)
df['수신합계'] = df[['요구불예금잔액', '거치식예금잔액', '적립식예금잔액']].sum(axis=1)

# 예대비율 계산 (0으로 나누는 오류 방지)
df['예대비율'] = df['여신합계'] / df['수신합계'].replace(0, np.nan)
# 업종별 평균 예대비율 집계
업종별_예대비율 = df.groupby('업종_대분류')['예대비율'].mean().sort_values(ascending=False)
```

```python
# 수신합계가 0인 경우는 예대비율 NaN으로 처리했지만, 극단값은 제거 필요
df_filtered = df[(df['예대비율'] < 100)]  # 예: 예대비율 100배 초과 제거 (상황에 맞게 조절)

# 다시 업종별 전담고객 여부 평균 예대비율 재계산
업종_전담고객별_예대비율_filtered = df_filtered.groupby(['업종_대분류', '전담고객여부'])['예대비율'].mean().unstack()

# 시각화
plt.figure(figsize=(12, 6))
업종_전담고객별_예대비율_filtered.plot(kind='bar', color=['lightgray', 'steelblue'])
plt.title('업종별 전담고객 여부에 따른 평균 예대비율')
plt.xlabel('업종 대분류')
plt.ylabel('예대비율 (배)')
plt.xticks(rotation=45, ha='right')
plt.legend(title='전담고객 여부')
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
