```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
import seaborn as sns
import matplotlib.pyplot as plt
plt.rcParams['font.family']='Malgun Gothic'# 맥 AppleGothic
plt.rcParams['axes.unicode_minus']=False
```

```python
df1 = pd.read_csv('../Datas/iM_data.csv',encoding='cp949')
df1
```

```python
df1.info()
```

```python
# ① 총 여신/예금 계산
df1['총여신잔액'] = df1['여신_운전자금대출잔액'] + df1['여신_시설자금대출잔액']
df1['총예금잔액'] = df1['거치식예금잔액'] + df1['적립식예금잔액']
df1['이자수익기여도'] = df1['총여신잔액'] - df1['총예금잔액']

# ② 업종_중분류 + 사업장_시도별 평균
grouped = df1.groupby(['업종_대분류', '사업장_시도'])[
    ['총여신잔액', '총예금잔액', '이자수익기여도']
].mean().reset_index()

# ③ 확인
grouped.head()
```

```python
# 주요 분석 컬럼
amount_cols = [
    '요구불예금잔액', '거치식예금잔액', '적립식예금잔액',
    '퇴직연금잔액', '신탁잔액', '수익증권잔액',
    '인터넷뱅킹거래금액', '스마트뱅킹거래금액', '폰뱅킹거래금액',
    'ATM거래금액', '자동이체금액',
    '신용카드사용금액', '체크카드사용금액', '창구거래금액',
    '외환_수출실적금액', '외환_수입실적금액'
]

# 그룹핑: 업종 + 시도 단위 평균
grouped = df1.groupby('업종_대분류')[amount_cols].mean().reset_index()
grouped.head()
```

```python
for col in amount_cols:
    temp = grouped[['업종_대분류', col]].sort_values(by=col, ascending=False)
    print(f'\n📌 {col} 상위 업종:')
    display(temp.head(5))  # Jupyter 환경일 경우
```

```python
# 예금 성격만 따로 집계
deposit_cols = ['요구불예금잔액', '거치식예금잔액', '적립식예금잔액']
deposit_group = df1.groupby('업종_대분류')[deposit_cols].mean().sort_values(by='요구불예금잔액', ascending=False)

# 시각화
deposit_group.plot(kind='bar', figsize=(14, 6))
plt.title('업종별 예금 잔액 평균')
plt.ylabel('금액 (백만원 단위)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
df1['디지털채널금액'] = (
    df1['인터넷뱅킹거래금액'] +
    df1['스마트뱅킹거래금액'] +
    df1['폰뱅킹거래금액']
)

digital_group = df1.groupby('업종_대분류')['디지털채널금액'].mean().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=digital_group.values, y=digital_group.index)
plt.title('업종별 디지털 채널 거래금액 평균')
plt.xlabel('금액 (백만원)')
plt.tight_layout()
plt.show()
```

```python
deposit_cols = [
    '요구불예금잔액', '거치식예금잔액', '적립식예금잔액',
    '퇴직연금잔액', '신탁잔액', '수익증권잔액'
]
transaction_cols = [
    '인터넷뱅킹거래금액', '스마트뱅킹거래금액', '폰뱅킹거래금액',
    'ATM거래금액', '자동이체금액',
    '신용카드사용금액', '체크카드사용금액', '창구거래금액'
]
fx_cols = [
    '외환_수출실적금액', '외환_수입실적금액'
]
loan_cols = [
    '여신_운전자금대출잔액', '여신_시설자금대출잔액'
]
```

```python
summary_df = df1.groupby('업종_대분류').agg({
    **{col: 'mean' for col in deposit_cols + loan_cols + transaction_cols + fx_cols}
}).reset_index()

# 항목군별 합계 컬럼 추가
summary_df['예금합계'] = summary_df[deposit_cols].sum(axis=1)
summary_df['여신합계'] = summary_df[loan_cols].sum(axis=1)
summary_df['거래합계'] = summary_df[transaction_cols].sum(axis=1)
summary_df['외환합계'] = summary_df[fx_cols].sum(axis=1)

# 보기 좋게 정리
final = summary_df[['업종_대분류', '예금합계', '여신합계', '거래합계', '외환합계']]
final = final.sort_values(by='여신합계', ascending=False)
display(final.head())
```

```python
# 이 값이 클수록 대출이 예금보다 많은 업종 = 은행 수익성이 높은 업종일 수 있습니다.
final['여신/예금 비율'] = final['여신합계'] / final['예금합계']
final.sort_values(by='여신/예금 비율', ascending=False).head()
```

```python
# 예금보다 대출 규모가 훨씬 큰 업종
top = final.sort_values(by='여신/예금 비율', ascending=False).head(10)

plt.figure(figsize=(10, 6))
sns.barplot(x='여신/예금 비율', y='업종_대분류', data=top)
plt.title('업종별 여신/예금 비율 상위 10개')
plt.tight_layout()
plt.show()
```

```python
final['여신+거래 합계'] = final['여신합계'] + final['거래합계']
final.sort_values(by='여신+거래 합계', ascending=False).head()
```

```python
# 전담고객 포함 구조
summary_by_customer = df1.groupby(['업종_대분류', '전담고객여부']).agg({
    **{col: 'mean' for col in deposit_cols + loan_cols + transaction_cols + fx_cols}
}).reset_index()

# 항목군별 합계
summary_by_customer['예금합계'] = summary_by_customer[deposit_cols].sum(axis=1)
summary_by_customer['여신합계'] = summary_by_customer[loan_cols].sum(axis=1)
summary_by_customer['여신/예금 비율'] = summary_by_customer['여신합계'] / summary_by_customer['예금합계']
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(
    data=summary_by_customer.sort_values(by='여신/예금 비율', ascending=False).head(10),
    x='여신/예금 비율', y='업종_대분류', hue='전담고객여부'
)
plt.title('업종별 여신/예금 비율 (전담고객 포함)')
plt.tight_layout()
plt.show()
```
