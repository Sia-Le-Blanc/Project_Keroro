```python
selected_cols = [
    '기준년월',
    '가명 식별자',
    '업종(중분류)',
    '설립일자',
    '종업원수',
    '자산총계',
    '유동자산',
    '비유동자산',
    '부채총계',
    '자기자본(납입자본금)',
    '이익잉여금',
    '유동부채',
    '비유동부채',
    '단기차입금',
    '차입금',
    '현금성자산',
    '매출채권',
    '매출액',
    '영업손익',
    '당기순이익',
    'EBITDA',
    '총자산증가율',
    '부채비율',
    '자기자본비율',
    '기준일로부터 향후 1년내 부도, 기업회생, 90일이상 금융연체, 대지급 등 발생여부(신용정보원 기준). 1: 향후1년내 부도발생, 0: 향후1년내 부도미발생'
]
df2_1 = df2[selected_cols].copy()
```

```python
# 타겟 컬럼
target_col = '기준일로부터 향후 1년내 부도, 기업회생, 90일이상 금융연체, 대지급 등 발생여부(신용정보원 기준). 1: 향후1년내 부도발생, 0: 향후1년내 부도미발생'

# 데이터 집계
target_counts = df2_1[target_col].value_counts().sort_index()
target_labels = ['미발생(0)', '발생(1)']

# 시각화
plt.figure(figsize=(6, 4))
sns.barplot(x=target_labels, y=target_counts.values, palette='Set2')
plt.title('향후 1년 내 부도 등 발생 여부 분포')
plt.xlabel('부도 발생 여부')
plt.ylabel('건수')
for i, count in enumerate(target_counts.values):
    plt.text(i, count + 200, f'{count:,}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
