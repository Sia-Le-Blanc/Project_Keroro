```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 필터: 건설업과 부동산업만 추출
df_focus = df[df['업종_대분류'].isin(['건설업', '부동산업'])].copy()

# 1. 업종별 평균 대출잔액 비교 (운전자금 vs 시설자금)
loan_amt = df_focus.groupby('업종_대분류')[['여신_운전자금대출잔액', '여신_시설자금대출잔액']].mean().reset_index()
loan_amt = loan_amt.melt(id_vars='업종_대분류', var_name='대출_유형', value_name='평균_대출잔액')

plt.figure(figsize=(8, 5))
sns.barplot(data=loan_amt, x='업종_대분류', y='평균_대출잔액', hue='대출_유형')
plt.title('업종별 평균 대출 잔액 비교')
plt.ylabel('평균 대출잔액')
plt.xticks(rotation=0)
plt.show()

# 3. 건설업 내부 대출 비중 시각화
construction = df_focus[df_focus['업종_대분류'] == '건설업']
con_agg = {
    '운전자금': construction['여신_운전자금대출잔액'].sum(),
    '시설자금': construction['여신_시설자금대출잔액'].sum()
}
plt.figure(figsize=(5, 5))
plt.pie(con_agg.values(), labels=con_agg.keys(), autopct='%1.1f%%', startangle=90)
plt.title('건설업 대출 유형 비중 (잔액 기준)')
plt.show()

# 4. 부동산업 내부 대출 비중 시각화
real_estate = df_focus[df_focus['업종_대분류'] == '부동산업']
re_agg = {
    '운전자금': real_estate['여신_운전자금대출잔액'].sum(),
    '시설자금': real_estate['여신_시설자금대출잔액'].sum()
}
plt.figure(figsize=(5, 5))
plt.pie(re_agg.values(), labels=re_agg.keys(), autopct='%1.1f%%', startangle=90)
plt.title('부동산업 대출 유형 비중 (잔액 기준)')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 분석할 좌수 컬럼
좌수_컬럼 = ['여신_운전자금대출좌수', '여신_시설자금대출좌수']

# 2. 건설업과 부동산업만 필터링
df_focus = df[df['업종_대분류'].isin(['건설업', '부동산업'])].copy()

# 3. 업종별 좌수 합계 계산
loan_cnt = df_focus.groupby('업종_대분류')[좌수_컬럼].sum().reset_index()

# 4. 시각화를 위한 long-form 변환
loan_cnt = loan_cnt.melt(id_vars='업종_대분류', var_name='대출_유형', value_name='총_건수')

# 5. 시각화
plt.figure(figsize=(8, 5))
sns.barplot(data=loan_cnt, x='업종_대분류', y='총_건수', hue='대출_유형')
plt.title('건설업 vs 부동산업 대출 건수(좌수) 비교')
plt.ylabel('총 건수')
plt.xlabel('업종')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
df['사업장_시도'].isnull().sum()  # → 17,000건 이상 존재

# 필터 전 결측 제거
df = df[df['사업장_시도'].notna()]
```

```python
# 대구광역시 + 부동산업 필터링
df_daegu_realestate = df[
    (df['사업장_시도'] == '대구광역시') &
    (df['업종_대분류'] == '부동산업')
].copy()

# 결과 확인
print(df_daegu_realestate.shape)
df_daegu_realestate
```

```python
# 시군구별 여신 잔액 평균 계산
grouped = df_daegu_realestate.groupby('사업장_시군구')[['여신_시설자금대출잔액']].sum().reset_index()

# 시각화
plt.figure(figsize=(12, 6))
grouped_melted = pd.melt(grouped, id_vars='사업장_시군구', var_name='대출유형', value_name='평균잔액')
sns.barplot(data=grouped_melted, x='사업장_시군구', y='평균잔액', hue='대출유형')
plt.title('대구광역시 부동산업 - 시군구별 평균 여신 잔액 비교')
plt.ylabel('평균 여신 잔액 (원)')
plt.xlabel('대구광역시 시군구')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 시군구별 시설자금대출좌수 평균 계산
loan_count = df_daegu_realestate.groupby('사업장_시군구')[['여신_시설자금대출좌수']].sum().reset_index()

# 시각화
plt.figure(figsize=(12, 6))
sns.barplot(data=loan_count, x='사업장_시군구', y='여신_시설자금대출좌수', palette='coolwarm')
plt.title('대구광역시 부동산업 - 시군구별 평균 시설자금대출 좌수')
plt.ylabel('평균 시설자금대출 좌수')
plt.xlabel('대구광역시 시군구')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 부동산업만 필터링
df_realestate = df[df['업종_대분류'] == '부동산업'].copy()

# 1. 시도별 평균 시설자금대출잔액 계산
avg_by_region = df_realestate.groupby('사업장_시도')['여신_시설자금대출잔액'].mean().dropna().sort_values(ascending=False).reset_index()

# 2. 시각화
plt.figure(figsize=(12, 6))
palette = ['red' if sido == '대구광역시' else 'lightgray' for sido in avg_by_region['사업장_시도']]

sns.barplot(data=avg_by_region, x='사업장_시도', y='여신_시설자금대출잔액', palette=palette)
plt.title('시도별 평균 시설자금대출잔액 (대구광역시 강조)')
plt.xlabel('사업장 시도')
plt.ylabel('평균 시설자금대출잔액 (원)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. 부동산업만 필터링
df_realestate = df[df['업종_대분류'] == '부동산업'].copy()

# 2. 시도별 여신_시설자금대출잔액 총합 계산
sum_by_region = (
    df_realestate
    .groupby('사업장_시도')['여신_시설자금대출잔액']
    .sum()
    .dropna()
    .reset_index()
)

# 2. 대구광역시 강조 색상 설정
palette = ['red' if sido == '대구광역시' else 'lightgray' for sido in sum_by_region['사업장_시도']]

# 3. 시각화
plt.figure(figsize=(12, 6))
sns.barplot(data=sum_by_region, x='사업장_시도', y='여신_시설자금대출잔액', palette=palette)
plt.title('시도별 총 시설자금대출잔액 (대구광역시 강조)')
plt.xlabel('사업장 시도')
plt.ylabel('총 시설자금대출잔액 (원)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
