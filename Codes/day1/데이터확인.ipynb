```python
name, cnt = np.unique(df.업종_대분류, return_counts=True)
```

```python
zipd = zip(name, cnt)
```

```python
l = []
for i in zipd:
    l.append(i)
```

```python
ar = np.array(l)
```

```python
my_df = pd.DataFrame(ar, columns=['개수', '업종명'])
```

```python
my_df.개수 = my_df.개수.astype(int)
```

```python
my_df.sort_values('개수', ascending=False)
```
